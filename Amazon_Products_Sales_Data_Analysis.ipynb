### **Step 4: Data cleaning**

```python
# The fields 'Unnamed: 0', 'image', and 'link' are irrelevant to our analysis
# To remove these fields

df_dropped = df.drop(['Unnamed: 0', 'image', 'link'], axis = 1)

df_dropped.head()
```

```python
# To rename the columns to make them look proper

df_dropped.columns = ['Name', 'Main Category', 'Sub-category', 'Ratings', 'No. Of Ratings', 'Discount Price', 'Actual Price']

df_dropped.head()
```

```python
# To check for total number of missing values in each column

df_dropped.isnull().sum()
```

```python
# To calculate the percentage of the total dataset that has missing values

print((df_dropped.isnull().sum().sum() / np.product(df_dropped.shape)) * 100)
```

```python
# Since the missing/NaN values amount to less than 25% (one-quarter) of the total dataset, I can safely remove these rows
# To drop all rows with NaN values or missing values

df_dropped_2 = df_dropped.dropna(axis=0)

df_dropped_2.head()
```

```python
# To check for total number of missing values in each column

df_dropped_2.isnull().sum()
```

```python
# Out of 551585 total rows, I have 340680 left after I removed missing/NaN values
# To see the datatypes of each column in the dataframe

df_dropped_2.dtypes
```

I can see that even the Ratings, No. Of Ratings, Discount Price, and Actual Price fields are Object columns

I have to change the datatypes to Decimal, Integer, Decimal, and Decimal respectively

```python
# To check for unique values of ratings

df_dropped_2.Ratings.unique()
```

I cannot convert Ratings field from Object to float due to presence of non-numeric data

I have to get rid of these non-numeric values, by using .str.contains method to filter them out then drop them

```python
# To match the rupee sign so that I can filter out irrelevant entries in the dataset

df_dropped_2[df_dropped_2['Ratings'].str.contains('₹')]
```

```python
df_dropped_2.shape
```

```python
# droping the filter for rupee sign for which I used .str.contains method to filter out

df_dropped_3 = df_dropped_2.drop(df_dropped_2[df_dropped_2['Ratings'].str.contains('₹')].index)

df_dropped_3.head()
```

```python
# checking again for unique values of ratings by using .unique method

df_dropped_3.Ratings.unique()
```

```python
# now let's filter out the Get & Free from this dataset using conditional filtering

df_dropped_3[(df_dropped_3.Ratings == 'Get') | (df_dropped_3.Ratings == 'FREE')]
```

```python
# After doing the conditional filtering for Get and free, now I can pass it into a new variable

df_dropped_4 = df_dropped_3[(df_dropped_3.Ratings != 'Get') & (df_dropped_3.Ratings != 'FREE')]

df_dropped_4.head()
```

```python
# Typecasting Ratings to Decimal to perform more operations

df_dropped_4['Ratings'] = df_dropped_4['Ratings'].astype('float64')

df_dropped_4.dtypes
```

```python
# checking for unique values for No of ratings

df_dropped_4['No. Of Ratings'].unique()
```

I cannot convert No. Of Ratings field from Object to Integer when the records have comma sign

So, I have to get rid of these comma signs, and replace them with empty string

```python
df_dropped_4['No. Of Ratings'] = df_dropped_4['No. Of Ratings'].str.replace(',','')

df_dropped_4.head()
```

```python
# removing all string values from a column using regex

df_dropped_4['No. Of Ratings'] = df_dropped_4['No. Of Ratings'].replace('[a-zA-Z]', np.nan, regex=True)

df_dropped_4.head()
```

```python
# Typecasting No. Of Ratings to Integer (first to float because the column seems to have a mixture of integer and decimal)

df_dropped_4['No. Of Ratings'] = df_dropped_4['No. Of Ratings'].astype('float64')

df_dropped_4['No. Of Ratings'] = df_dropped_4['No. Of Ratings'].astype('int64')

df_dropped_4.dtypes
```

```python
# checking for unique values of discount price

df_dropped_4['Discount Price'].unique()
```

```python
# I cannot convert Discount Price field from Object to Integer/Float when the records have comma sign and the rupee sign
# I have to get rid of these signs, and replace them with empty string

df_dropped_4['Discount Price'] = df_dropped_4['Discount Price'].str.replace(',','')

df_dropped_4['Discount Price'] = df_dropped_4['Discount Price'].str.replace('₹', '')
```

```python
# Typecasting Discount Price to Decimal to perform more operations

df_dropped_4['Discount Price'] = df_dropped_4['Discount Price'].astype('float64')

df_dropped_4.dtypes
```

```python
# checking for unique values of Actual Price column

df_dropped_4['Actual Price'].unique()
```

```python
# I cannot convert Actual Price field from Object to Integer/Float when the records have comma sign and the rupee sign
# I have to get rid of these signs, and replace them with empty string

df_dropped_4['Actual Price'] = df_dropped_4['Actual Price'].str.replace(',','')

df_dropped_4['Actual Price'] = df_dropped_4['Actual Price'].str.replace('₹','')
```

```python
# Typecasting Actual Price to Integer/float to perform more operations

df_dropped_4['Actual Price'] = df_dropped_4['Actual Price'].astype('float64')

df_dropped_4.dtypes
```

```python
# checking for duplicated values in the dataset

df_dropped_4.duplicated().any()
```

```python
# droping the duplicates in the dataset and passing it into a new variable

df_dropped_4 = df_dropped_4.drop_duplicates()
```

```python
# After droping the duplicates, I can confirm if their is still duplicates
# As I can see below it's False, meaning their is no more duplicates in the dataset

df_dropped_4.duplicated().any()
```

```python
# looking at the cleaned dataset

df_dropped_5 = df_dropped_4

df_dropped_5
```
